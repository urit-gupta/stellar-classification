```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#for consistency
plt.style.use('bmh')

#read data
star_data = pd.read_csv('star_classification.csv')

#list important features
#r = value of red filter of object
#i = value of near-infrared filter of object
#redshift = a measure of how far the object is - more the redshift, more the distance
imp_features = ['r', 'i', 'redshift']

#keep only those features and the labels
star_data = star_data[imp_features + ['class']]

star_data.head()
```

```python
#check how many of each label are present
star_data['class'].value_counts()
```

```python
#initial distributions of features
star_data.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=9);
```

```python
#analyze relationships between features
sns.pairplot(data=star_data, vars=imp_features)
plt.show()
```

```python
#view numeric correlations between features
print(star_data[imp_features].corr())

sns.heatmap(star_data[imp_features].corr(), annot=True, cmap='Reds')
plt.show()
```

```python
#analyze relationships of red values with classes
sns.boxplot(star_data['class'], star_data['r'])
plt.show()
```

```python
#analyze relationships of near-infrared values with classes
sns.boxplot(star_data['class'], star_data['i'])
plt.show()
```

```python
#analyze relationships of redshift values with classes
sns.boxplot(star_data['class'], star_data['redshift'])
plt.show()
```

```python
#convert labels for analysis of clustering algorithms
true_y = []
for label in star_data['class']:
    true_y.append(0 if label == 'GALAXY' else (1 if label == 'QSO' else 2))

#remove the labels and keep only the features for clustering
X = star_data[imp_features]
y = star_data['class']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#prepare train & test split for features & labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

#scale both the train & test features for faster convergence
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

#setup for graphing mean classification error over k-values
error_x, error_y = [], []

#track the minimum mean error and associated k-value
min_error, k_with_min_error = 1.01, 0

#classify using KNN with k-values from 1 to 42
for k in range(1, 43):
    #the error graph's x-axis will have k-values
    error_x.append(k)
    
    #create a KNN classifier with k-value
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    
    #calculate mean error for current k-value
    error = np.mean(y_pred != y_test)
    
    #the error graph's y axis will have mean errors for each k-value
    error_y.append(error)
    
    #update minimum error and associated k-value if necessary
    min_error, k_with_min_error = (error, k) if error < min_error else (min_error, k_with_min_error)

plt.scatter(error_x, error_y)
plt.title('Error Rate / K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error in KNN')
plt.show()

print('Minimum Mean Error: {:.3f}'.format(min_error))
print('Associated K-Value: {}'.format(k_with_min_error))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

#use KNN with the k-value that produces minimum mean error
classifier = KNeighborsClassifier(n_neighbors=k_with_min_error)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

#plot confusion matrix of true and predicted labels
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Reds')
plt.title('Confusion Matrix of K-NN Results')
plt.xlabel('True Labels')
plt.ylabel('Predicted Labels')
plt.show()

#print classification report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score

#retrieve cross-validation scores
scores = cross_val_score(classifier, X, y, cv=5)

#plot cross-validation scores
plt.plot(range(1, 6), scores, marker='o', markersize=10)
plt.title('Cross-Validation Scores for KNN')
plt.xlabel('CV Split ID')
plt.ylabel('Score')
plt.xlim([0.5, 5.5])
plt.ylim([0.92, 1.0])
plt.show()

print('Mean CV score: {:.3f}'.format(np.mean(scores)))
```

```python
from sklearn.svm import SVC

#poly and rbf take waaaaaaaaaaaaaaaaaaaaa-KERNEL HAS DIED
kernel = 'linear'

#create linear Support Vector Machine Classifier
svclassifier = SVC(kernel=kernel)
svclassifier.fit(X_train, y_train)

y_pred = svclassifier.predict(X_test)

#plot confusion matrix of true and predicted labels
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Reds')
plt.title('Confusion Matrix of SVM-' + kernel + ' Results')
plt.xlabel('True Labels')
plt.ylabel('Predicted Labels')
plt.show()

#print classification report
print(classification_report(y_test, y_pred))
```

```python
#retrieve cross-validation scores
scores = cross_val_score(svclassifier, X, y, cv=5)

#plot cross-validation scores
plt.plot(range(1, 6), scores, marker='o', markersize=10)
plt.title('Cross-Validation Scores for SVM-' + kernel)
plt.xlabel('CV Split ID')
plt.ylabel('Score')
plt.xlim([0.5, 5.5])
plt.ylim([0.92, 1.0])
plt.show()

print('Mean CV score: {:.3f}'.format(np.mean(scores)))
```
